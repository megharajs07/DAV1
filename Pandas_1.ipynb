### Rename columns

We can rename columns by
1.   passing the dictionary with old_name: new_name pair
2.   specifying axis = 1

```python
df.columns
```

```python
dict1 = {"population": "countofpeople", "life_exp": "life_expectancy"}
```

```python
df.rename(columns = dict1) # rename column index
```

```python
df.rename(dict1, axis = 1) # rename column index
```
