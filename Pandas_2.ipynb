- Thats the end of lexicographical sorting. Now we'll learn about `pd.concat`

````python
user = pd.DataFrame({"userid":[1, 2, 3], "name":["sharadh", "shahid", "khusalli"]})
user
````

````python
msgs = pd.DataFrame({"userid":[1, 1, 2, 4], "msg":['hmm', "acha", "theek hai", "nice"]})
msgs
````

````python
pd.concat([user, msgs], axis = 0)
````

````python
pd.concat([user, msgs], axis = 1)
````

#### Merge:

`pd.merge`

````python
user
````

````python
msgs
````

````python
user.merge(msgs, on = 'userid')
````

````python
pd.concat([user, msgs], axis =1)
````

````python
pd.merge(user, msgs, left_on = 'userid', right_on='userid')
````

````python
msgs = pd.DataFrame({"id":[1, 1, 2, 4], "msg":['hmm', "acha", "theek hai", "nice"]})
msgs
````

````python
pd.merge(user, msgs, left_on = 'userid', right_on='id')
````

````python
pd.merge(user, msgs, left_on = 'userid', right_on = 'id', how = 'left')
````
