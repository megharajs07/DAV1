# High Risk directors:
- if a movie has higher budget than avg for that director => high risk director

```python
def risky(x):
  x['risky'] = x['budget'] - x['budget'].mean() > 0
  return x
```

```python
data['budget'].mean()
```

```python
data.groupby('director_name').apply(risky)
```

```python
data = data.groupby('director_name', group_keys = False).apply(risky)
```

```python
data['risky'].value_counts()
```

```python
data[data['risky'] == True]
```

# Most Productive Directors

- More work, more productive, more movies, for the timeperiod

```python
data['director_name'].value_counts()
```

```python
data.groupby('director_name')['year'].aggregate(['min','max'])
```

```python
data_agg = data.groupby('director_name').aggregate({'year' : ['min','max'], 'title':['count']})
```

```python
data_agg
```

```python
data_agg.columns
```

```python
data_agg.head()
```

```python
data_agg.columns = ['year_min', 'year_max', 'title_count']
data_agg
```

```python
data_agg.reset_index(inplace = True)
data_agg
```

```python
data_agg['active_years'] = data_agg['year_max'] - data_agg['year_min']
data_agg
```

```python
data_agg['movies_per_year'] = data_agg['title_count'] / data_agg['active_years']

data_agg
```

```python
data_agg.sort_values(by = 'movies_per_year', ascending = False,inplace = True)
data_agg
```
